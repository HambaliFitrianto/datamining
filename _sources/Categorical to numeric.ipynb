# Categorical to Numeric

```python
import numpy as np
import pandas as pd
```

```python
dataset_url = 'https://raw.githubusercontent.com/HambaliFitrianto/datamining/main/credit_score.csv'
data = pd.read_csv(dataset_url)
```

```python
data.head()
```

```python
data_asli = data.drop(columns=['kpr_aktif', 'rata_rata_overdue'])
```

```python
data_kpr = data[["kpr_aktif"]]
```

```python
data_kpr.head(25)
```

```python
kpr_aktif = pd.get_dummies(data_kpr)
```

```python
kpr_aktif
```

```python
data_rata_rata_overdue = data[["rata_rata_overdue"]]
```

```python
data_rata_rata_overdue.head(25)
```

```python
rata_rata_overdue = pd.get_dummies(data_rata_rata_overdue)
```

```python
rata_rata_overdue
```

```python
data_olah = pd.concat([kpr_aktif, rata_rata_overdue], axis=1)
```

```python
data_hasil = pd.concat([data_asli, data_olah], axis=1)
```
