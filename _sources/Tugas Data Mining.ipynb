```python
import pandas as pd
import numpy as np
import math 
```

```python
dataset_url="https://raw.githubusercontent.com/HambaliFitrianto/datamining/main/data-kejadian-bencana-banjir-di-provinsi-dki-jakarta-tahun-2017-bulan-desember.csv"
p=pd.read_csv(dataset_url, index_col=0)
```

```python
p
```

```python
def diss(x,y):
  a = 4
  b = 0
  for col in ["kecamatan", "kelurahan", "rw", "jumlah_terdampak_rw", "jumlah_terdampak_rt",	"jumlah_terdampak_kk", "jumlah_terdampak_jiwa",	"ketinggian_air",	"tanggal_kejadian",	"lama_genangan",	"jumlah_meninggal", "jumlah_hilang",	"jumlah_luka_berat",	"jumlah_luka_ringan",	"jumlah_pengungsi_tertinggi",	"jumlah_tempat_pengungsian",	"nilai_kerugian", "kota_administrasi"	]:
    if p[col][x] == p[col][y]:
      b+=1
    return (a-b)/a
```

```python
print(diss(1,1))
```

```python
p.describe()
```

```python
#normalisasi jumlah_terdampak_rw
std_jtw = 1.280024
mean_jtw = 1.961538
for col1 in ["jumlah_terdampak_rw"]:
    p1 = p[col1]
    hasil1 = (p1-mean_jtw)/std_jtw
    print(hasil1)

#normalisasi jumlah_terdampak_rt
std_jtr = 2.324783	
mean_jtr = 3.366667
for col2 in ["jumlah_terdampak_rt"]:
    p2 = p[col2]
    hasil2 = (p2-mean_jtr)/std_jtr
    print(hasil2)
```

```python
# Menghitung jarak data Numeric dengan cara Euclidean
def num_p(a,b,c):
    result = 0
    for col in [hasil1, hasil2]:
        result = math.pow(abs(col[a] - col[b]), c)
    return result**(1/c)
```

```python
print(num_p(1,2,2))
print(num_p(1,3,2))
print(num_p(1,4,2))
```

**Similarity**

```python
print("Jarak d(1,2) =", diss(1,2)+num_p(1,2,2))
```

```python
print("Jarak d(1,2) =", diss(1,3)+num_p(1,3,3))
```

```python
print("Jarak d(1,2) =", diss(1,4)+num_p(1,4,4))
```

```python
# Feature matrix in a object named X
X = iris.data
# response vector in a object named y
y = iris.target
```

```python
print(X.shape)
print(y.shape)
```

Train the Model

```python
# splitting the data into training and test sets (80:20)
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
#shape of train and test objects
print(X_train.shape)
print(X_test.shape)
```

```python
# shape of new y objects
print(y_train.shape)
print(y_test.shape)
```

```python
#import the KNeighborsClassifier class from sklearn
from sklearn.neighbors import KNeighborsClassifier

#import metrics model to check the accuracy 
from sklearn import metrics
#Try running from k=1 through 25 and record testing accuracy
k_range = range(1,26)
scores = {}
scores_list = []
for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train,y_train)
        y_pred=knn.predict(X_test)
        scores[k] = metrics.accuracy_score(y_test,y_pred)
        scores_list.append(metrics.accuracy_score(y_test,y_pred))
```

```python
#Testing accuracy for each value of K
scores
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

#plot the relationship between K and the testing accuracy
plt.plot(k_range,scores_list)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

K values with 3 to 19 has the same accuracy which is 96.66, so we can use any one value from that, i choose K as 5 and train the model with full training data

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,y)
```

```python
#0 = setosa, 1=versicolor, 2=virginica
classes = {0:'setosa',1:'versicolor',2:'virginica'}

#Making prediction on some unseen data 
#predict for the below two random observations
x_new = [[3,4,5,2],
         [5,4,2,2]]
y_predict = knn.predict(x_new)

print(classes[y_predict[0]])
print(classes[y_predict[1]])
```

# Tugas 4 : Naive Bayes Classifer

```python
# Naive Bayes Classification

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('iris.csv')
```

```python
#looking at the first 5 values of the dataset
dataset.head()
```

```python
%matplotlib inline
img=mpimg.imread('iris_types.jpg')
plt.figure(figsize=(20,40))
plt.axis('off')
plt.imshow(img)
```

```python
#Spliting the dataset in independent and dependent variables
X = dataset.iloc[:,:4].values
y = dataset['species'].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 82)
```

```python
# Feature Scaling to bring the variable in a single scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting Naive Bayes Classification to the Training set with linear kernel
from sklearn.naive_bayes import GaussianNB
nvclassifier = GaussianNB()
nvclassifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = nvclassifier.predict(X_test)
print(y_pred)
```

```python
#lets see the actual and predicted value side by side
y_compare = np.vstack((y_test,y_pred)).T
#actual value on the left side and predicted value on the right hand side
#printing the top 5 values
y_compare[:5,:]
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#finding accuracy from the confusion matrix.
a = cm.shape
corrPred = 0
falsePred = 0

for row in range(a[0]):
    for c in range(a[1]):
        if row == c:
            corrPred +=cm[row,c]
        else:
            falsePred += cm[row,c]
print('Correct predictions: ', corrPred)
print('False predictions', falsePred)
print ('\n\nAccuracy of the Naive Bayes Clasification is: ', corrPred/(cm.sum()))   
```
