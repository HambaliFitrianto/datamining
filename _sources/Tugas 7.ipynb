# Tugas 7 (Studi Kasus Data Kredit)

```python
import numpy as np
import pandas as pd
```

```python
dataset_url = 'https://raw.githubusercontent.com/HambaliFitrianto/datamining/main/credit_score.csv'
data = pd.read_csv(dataset_url)
```

```python
data.head()
```

```python
data_asli = data.drop(columns=['kpr_aktif', 'rata_rata_overdue'])
```

```python
data_kpr = data[["kpr_aktif"]]
```

```python
data_kpr.head(25)
```

```python
kpr_aktif = pd.get_dummies(data_kpr)
```

```python
kpr_aktif
```

```python
data_rata_rata_overdue = data[["rata_rata_overdue"]]
```

```python
data_rata_rata_overdue.head(25)
```

```python
rata_rata_overdue = pd.get_dummies(data_rata_rata_overdue)
```

```python
rata_rata_overdue
```

```python
data_olah = pd.concat([kpr_aktif, rata_rata_overdue], axis=1)
```

```python
data_hasil = pd.concat([data_asli, data_olah], axis=1)
```

```python
data_hasil
```

**Normalisasi Data**

```python
normalisasi_data = data_hasil.drop(columns=['kode_kontrak', 'risk_rating'])
normalisasi_data
```

```python
from sklearn.preprocessing import MinMaxScaler
# scale features
scaler = MinMaxScaler()
model =scaler.fit(normalisasi_data)
scaled_data=model.transform(normalisasi_data)
# print scaled features
print(scaled_data)
```

```python
nama_kolom = normalisasi_data.columns.values
data_MinMax = pd.DataFrame(scaled_data, columns=nama_kolom)
data_MinMax
```

```python
# Min Max Scale dengan Min = 1 dan Max =2
scaler = MinMaxScaler(feature_range=(1,2))
model =scaler.fit(normalisasi_data)
scaled_data2=model.transform(normalisasi_data)
# print scaled features
print(scaled_data2)
```

```python
data_MinMax2 = pd.DataFrame(scaled_data2, columns=normalisasi_data.columns.values)
data_MinMax2
```

```python
# Z Score atau StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
model = (scaler.fit(normalisasi_data))
data_mean = (scaler.mean_)
scale_data = (scaler.transform(normalisasi_data))
print(scale_data)
```

```python
data_ZScale = pd.DataFrame(scale_data, columns=normalisasi_data.columns.values)
data_ZScale
```
